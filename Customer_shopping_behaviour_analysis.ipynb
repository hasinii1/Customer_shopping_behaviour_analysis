## Customer Shopping Behaviour Analysis

### Problem Statement:
A leading retail company wants to better understand its customers' shopping behavior in order to improve sales, customer satisfaction, and long-term loyalty. The management team has noticed changes in purchasing patterns across demographics, product categories, and sales channels
(online vs. offline). They are particularly interested in uncovering which factors, such as discounts reviews, seasons, or payment preferences, drive consumer decisions and repeat purchases.
You are tasked with analyzing the company's consumer behavior dataset to answer the following overarching business question:

"How can the company leverage consumer shopping data to identify trends, improve
customer engagement, and optimize marketing and product strategies?"

```python
import pandas as pd
```

```python
df = pd.read_csv('customer_shopping_behavior.csv')
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})

df.columns
```

```python
# create a column age_group
labels = ['Young adult' , 'Adult' , 'Middle-aged' , 'Senior' ]
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)

df[['age', 'age_group']].head(9)
```

```python
# create another col purchase_frequency_das-->convert the textual frequencies to numbers, basically no of days between purchases

frequency_mapping ={
    'Fortnighty':14,
    'Weekly':7,
    'Bi-Weekly':14,
    'Monthly':30,
    'Quarterly':90,
    'Anually':365,
    'Every 3 Months':90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)    

df[['purchase_frequency_days', 'frequency_of_purchases']].head(9)
```

```python
df[['discount_applied', 'promo_code_used']].head(5)
```

```python
(df['discount_applied']==df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used', axis=1)
df.columns
```

```python
print("Total Customers :", len(df))
print("Total Revenue   :", df['purchase_amount'].sum())
print("Average Spend   :", round(df['purchase_amount'].mean(),2))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Revenue by category
gender_sales = df.groupby('category')['purchase_amount'].sum()

plt.figure(figsize=(4,4))

plt.pie(
    gender_sales,
    labels=gender_sales.index,
    autopct='%1.1f%%',
    wedgeprops=dict(width=0.4)
)

plt.title("Revenue Contribution")
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.histplot(
    df['purchase_amount'],
    bins=20,
    kde=True
)
plt.title("Purchase Amount Distribution")
plt.xlabel("Purchase Amount")
plt.show()
```

```python
#Seasional Revenue
season_sales = (
    df.groupby('season')
    ['purchase_amount']
    .sum()
)
plt.figure(figsize=(4,4))
sns.lineplot(
    x=season_sales.index,
    y=season_sales.values,
    marker='o'
)
plt.title("Revenue Trend by Season")
plt.show()
```

```python
numeric_cols = df.select_dtypes(include='number')
corr = numeric_cols.corr()
plt.figure(figsize=(8,5))
sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(
    df[
        [
            'age',
            'purchase_amount',
            'review_rating',
            'previous_purchases'
        ]
    ]
)
plt.show()
```
